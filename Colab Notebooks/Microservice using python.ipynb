````python
# prompt: what is use of flask, flask-restful and  redis using python in simple words

# Flask:  A lightweight web framework.  Think of it as the engine that lets you build APIs (ways for different parts of your system to talk to each other).  You use it to define routes (URLs) and what happens when someone visits them.

# Flask-RESTful:  Makes it easier to create REST APIs with Flask.  REST APIs are a standard way for applications to communicate.  Flask-RESTful helps organize your API code.

# Redis: An in-memory data store. It's super fast for storing and retrieving data. In your microservices, you might use Redis to share data between services (like storing cart information so it's available to both the Cart Service and Order Service) or for caching (storing frequently used data so you don't have to look it up every time).
````
