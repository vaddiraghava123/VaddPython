```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten, Dense
```

```python
model = Sequential()
```

```python
# input layer, starts flatten 28 * 28  =  1 Dimensional

model.add(Flatten(input_shape = (28,28))) # 28 * 28 = 784 input
```

```python
# One Hidden layer
# how many neurons wants to use : 128
# which activation function wants to use " relu"
# bcz it avoid vanish graidents problem

model.add(Dense(units = 128, activation = "relu"))  # 128 :  784 * 128+128 *(1bias)
```

```python
# Output layer

# how many classes are there =10
# how many neurons : 10
# activation : softmax
model.add(Dense(10, activation = 'softmax')) # 18 is attached with 10 neurons

```

**Hidden layer**

-  There is no rule choose no. of hidden layers and neurons in the hidden layer

- Its completly depends on problem

- If problem is complex will use  3+ hidden layers 256 or 512 or 1024 neurons

- But more hidden layers and more neurons model become overfit

-  Then use regularization methods or Drop out method



**Drop out**

- Drop out method is the way to drop or deactivate neurons based on probability

-  will drop the neurons based on probability

- model.add(Dropout =0.5) indicates 50% chance that neuron might be drop

- Drop out will happend only on Training time .. not infernece




**Output layer**

-  Multi class mean multiple neurons, soft max

- Biclass means single neurons, sigmoid

```python
from tensorflow import keras
from keras.layers import Flatten, Dense, Dropout
from keras.models import Sequential

model = Sequential()
model.add(Flatten(input_shape = (28,28))) # 28 * 28 = 784 input
model.add(Dense(units = 128, activation = "relu"))  # 128 :  784 * 128+128 *(1bias)

# Add dropout
model.add(Dropout(0.5))
model.add(Dense(10, activation = 'softmax')) # 18 is attached with 10 neurons
```

-  When we use drop the output will be shrink

- To avoid this we will scale the remaining neurons output by 1/ 1- p

-  While training, we are scaling

-  While inference (testing) will use all the neurons

**Trainable params**

-  Updated by model during back propagation

- Trainable parameters mean weights and bias for each layer

**Non-trainable parameter**

- Not updated by model during back propagation

- The best example is Batch Normalization

- We already know that Normalization mean scaling the value to make all values under one scale

- Different activation values we need to scale it and this method called as Batch Normalization

**Batch Normalization**

-  model.add(BatchNormalization())

-  Interanl work of BatchNormalization

      -  Mean and variance z = x-u/sigma

      - new value = gamma * scaled value + beta

      -   gamma refers scaling factor

      -  beta refers shifting factor
